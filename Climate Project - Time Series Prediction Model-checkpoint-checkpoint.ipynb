## Northeast Data

```python
filename1 = r"C:\Users\hopey\OneDrive - University at Buffalo\Forecast NE.csv" #LOADING DATA
NEdata2001 = pd.read_csv(filename1)
filename2 = r"C:\Users\hopey\OneDrive - University at Buffalo\2021 MAX TEMP NE.csv"
NEdata2021 = pd.read_csv(filename2)
```

```python
Temp2001_NE = NEdata2001['Forecast(Value)'][240:-1] #last 12 pieces of forecasted data
Date2001_NE = NEdata2001['Date'] #month and year mmm-yy of data
Temp2021_NE = NEdata2021['Value'] #actual temperature data of 2021
Date2021 = NEdata2021['Date'] #corresponding month of 2021
```

```python
Temp2021_NE
```

```python
Temp2001_NE
```

```python
ForecastNE = [] #create list to contain the forecasted temperatur
ActualNE = [] #create list to contain the actual observed tempe
errorNE = [] #create list to contain the error of the forescas
#to the actual
for x,y in zip(Temp2001_NE,Temp2021_NE):
    ForecastNE.append(x) #adds each months forecasted temp to Forecast lis
    ActualNE.append(y) #adds each months actual temp to error list
    errorNE.append(abs(x-y)) #adds each months magnitude of error in degrees f
#list
```

```python
errorNE
```

```python
st_dev = [] #creates a standard deviation list for each region
st_dev.append(st.stdev(errorNE)) #adds the Northeastern error standard devation to t
means_lst = [] #creates a mean list for each region
means_lst.append(st.mean(errorNE)) #adds the Northeastern error mean to the list
Data_Sources = [] #creates a Data Sources list of each region
Data_Sources.append('Northeast') #adds the region to the list
```

```python
plt.plot(ForecastNE, 'or-', ActualNE, 'ob-')
plt.xlabel('Months of the year')
plt.ylabel('Temperature(Farenheit)')
plt.title('Monthly Maximum Temperature in the Northeastern U.S.')
plt.legend(['Forecast Month','Actual data'])
```
